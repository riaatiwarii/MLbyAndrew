```python
def compute_cost_logistic(X,y,w,b):
    m=len(X)
    cost=0.0
    for i in range(m):
        z_i=np.dot(X[i],w)+b
        f_wb_i=sigmoid(z_i)
        cost+= -y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1-f_wb_i)
    cost/=m
    return cost
```
